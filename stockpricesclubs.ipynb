```python
from yahooquery import Ticker
import pandas as pd
import time

# List of stock tickers for all clubs
tickers = [
    "JUVE.MI", "BVB.DE", "MANU", "GSRAY.IS", "FENER.IS",
    "BJKAS.IS", "PARKEN.CO", "SSL.MI", "SLBEN.LS", "FCP.LS",
    "SCP.LS", "SIF.CO", "BIF.CO", "CCP.L", "AJAX.AS",'AGF-B.CO', 'AAB.CO', 'BIF.CO', 'TSPOR.IS'
]

# Create an Excel writer to store multiple sheets
excel_filename = "Football_Clubs_Stock_Data.xlsx"

with pd.ExcelWriter(excel_filename, engine="openpyxl") as writer:
    for ticker in tickers:
        try:
            print(f"📊 Fetching data for {ticker}...")

            # Fetch stock data
            stock = Ticker(ticker)
            data = stock.history(start="1995-01-01", end="2025-02-24")

            # Fetch the latest available shares outstanding
            shares_outstanding = stock.key_stats.get(ticker, {}).get("sharesOutstanding", "N/A")

            # Add shares outstanding as a new column
            data["shares_outstanding"] = shares_outstanding

            # Save each stock's data in a separate sheet
            data.to_excel(writer, sheet_name=ticker)

            print(f"✅ Data fetched and saved for {ticker}")

            # Avoid hitting rate limits
            time.sleep(2)

        except Exception as e:
            print(f"❌ Failed to fetch data for {ticker}: {e}")

print(f"\n📊 All stock data saved in {excel_filename}")


```
